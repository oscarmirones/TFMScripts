Notebook donde se pretende agrupar todos los datos grupos en observaciones vectoriales. Cada observación será un vector/array 6-dimensional con las siguientes variables:

* Precipitación
* Radio
* Ángulo
* Presión Mínima
* Latitud
* SST

```python
import xarray as xr
import numpy as np
```

```python
import time
```

```python
start_time = time.time()

pmin = []
sst = []
lat = []
precip = []
dist = []
ang = []
for i in range(0,1636):
    
    ciclon = xr.open_dataarray('All_cyclones/precip_ciclon_all'+str(i)+'.nc4')
    cic = xr.open_dataset('All_cyclones/ciclon_all'+str(i)+'.nc4')
    if ciclon.shape[0] == 0:
        continue
    dist.append(xr.open_dataarray('All_cyclones/dist_ciclon_all'+str(i)+'.nc4').values.ravel())
    if cic.basin[0] == 1 or cic.basin[0] == 4 or cic.basin[0] == 6 or cic.basin[0] == 9 or cic.basin[0]==10:
        ang.append(np.radians(360) -xr.open_dataarray('All_cyclones/ang_ciclon_all'+str(i)+'.nc4').values.ravel())
    else:
        ang.append(xr.open_dataarray('All_cyclones/ang_ciclon_all'+str(i)+'.nc4').values.ravel())
    precip.append(xr.open_dataarray('All_cyclones/precip_ciclon_all'+str(i)+'.nc4').values.ravel())
    for j in range(0,ciclon.shape[0]):
        pmin.append(np.repeat(cic.minPressure[j].values,ciclon.shape[1]*ciclon.shape[2]))
        sst.append(np.repeat(cic.sst[j].values,ciclon.shape[1]*ciclon.shape[2]))
        lat.append(np.repeat(cic.latitude[j].values,ciclon.shape[1]*ciclon.shape[2]))
    
    print('cylone'+str(i)+'done!')
print('Time : ',time.time()-start_time)
```

```python
xr.Dataset({'precipitation': np.hstack(precip),
            'radius': np.hstack(dist),
            'theta': np.hstack(ang),
            'minPressure': np.array(pmin).ravel(), 'latitude':np.array(lat).ravel(), 'sst': np.array(sst).ravel() }).to_netcdf('raw_data2.nc4')
```
